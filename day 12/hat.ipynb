```python
import random

red_hats = 1
blue_hats = 1

print(f"Initial State: {red_hats} Red, {blue_hats} Blue\n")

for i in range(1, 101):
    total_hats = red_hats + blue_hats
    
    # Calculate the exact probability of picking a red hat at this moment
    probability_red = red_hats / total_hats
    
    # Draw a random number between 0 and 1
    if random.random() < probability_red:
        red_hats += 1
        picked = "Red"
        added = "Red"
    else:
        blue_hats += 1
        picked = "Blue"
        added = "Blue"
        
    print(f"Step {i}: Picked {picked} | Added {added} | Count: {red_hats} Red, {blue_hats} Blue")

print(f"\nFinal Count: {red_hats} Red Hats | {blue_hats} Blue Hats ({red_hats + blue_hats} Total)")
```

```python
import random

red_hats = 1
blue_hats = 1

print(f"Initial State: {red_hats} Red, {blue_hats} Blue\n")

for i in range(1, 101):
    total_hats = red_hats + blue_hats
    
    # Calculate the exact probability of picking a red hat at this moment
    probability_red = red_hats / total_hats
    
    # Draw a random number between 0 and 1
    if random.random() < probability_red:
        blue_hats += 1
        picked = "Red"
        added = "Blue"
    else:
        red_hats += 1
        picked = "Blue"
        added = "Red"
        
    print(f"Step {i}: Picked {picked} | Added {added} | Count: {red_hats} Red, {blue_hats} Blue")

print(f"\nFinal Count: {red_hats} Red Hats | {blue_hats} Blue Hats ({red_hats + blue_hats} Total)")
```
