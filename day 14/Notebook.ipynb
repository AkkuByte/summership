gcd code with steps


```python
def gcd(a, b):
    steps = 0
    while b:
        steps += 1
        print(f"Step {steps}: a = {a}, b = {b}")
        a, b = b, a % b
    print(f"GCD is {a} and it took {steps} steps.")
    return a    
# Example usage:
# take two numbers as input
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
gcd(num1, num2) 
```

gcd most steps 

```python
def gcd_steps(a, b):
    steps = 0
    while b:
        a, b = b, a % b
        steps += 1
    return steps

max_steps = 0
pairs = []

for a in range(10, 100):
    for b in range(10, 100):
        steps = gcd_steps(a, b)

        if steps > max_steps:
            max_steps = steps
            pairs = [(a, b)]
        elif steps == max_steps:
            pairs.append((a, b))

print("Maximum steps:", max_steps)
print("Pairs achieving maximum steps:")

for a, b in pairs:
    print(a, b)
```

for three digit 

```python
def gcd_steps(a, b):
    steps = 0
    while b:
        a, b = b, a % b
        steps += 1
    return steps

max_steps = 0
pairs = []

for a in range(100, 1000):
    for b in range(a + 1, 1000):  # avoid duplicate pairs
        steps = gcd_steps(a, b)

        if steps > max_steps:
            max_steps = steps
            pairs = [(a, b)]
        elif steps == max_steps:
            pairs.append((a, b))

print("Maximum steps:", max_steps)
print("Pairs achieving maximum steps:")

for a, b in pairs:
    print(a, b)
```

fibbonacci upto 200 terms 

```python
fib = [0, 1]

while len(fib) < 200:
    fib.append(fib[-1] + fib[-2])

for position, value in enumerate(fib, start=1):
    print(position, value)
```
